# Massive Table

```python
#Importing Massive Table
df=pd.read_csv("../../raw_data/redundant_data/Iron Ore Price/CMO-Hist.csv")
df['Year']=pd.to_datetime(df['Year'], format= '%YM%m')

#Importing BDI
bdi=pd.read_csv('../../raw_data/data/BDI/cleaned_weekly_BDI.csv')
bdi['time']=pd.to_datetime(bdi['time'],format='%Y-%m-%d')

#Creating Copies to merge then view their relation
bdi_copy=bdi.groupby(bdi['time'].dt.year).mean()
df_copy= df.apply(pd.to_numeric, errors='coerce')
df_copy=df_copy.groupby(df['Year'].dt.year).mean()
df_copy=df_copy.join(bdi_copy, how='left')
df_copy=df_copy[df_copy.index>1995]
```

```python
#Seeing the correlation between the features
import seaborn as sns
plt.figure(figsize=(20,20))
# calculate the correlation matrix
corr = df_copy.corr()
print(corr[['close']])
# plot the heatmap
sns.heatmap(corr, cmap="Blues")
```

```python
#Setting up df to use 
df.set_index('Year', inplace=True)
df= df.apply(pd.to_numeric, errors='coerce')
df=df[df.index>'1995-01-01']
```

**These are the columns that have the greatest correlation with BDI:** ['Natural_gas,_US','Shrimps,_Mexican','Barley','Soybean_oil','Platinum','Nickel','Aluminum','Tobacco,_US_import_u.v.','Orange','Sugar,_EU']

```python
plt.figure(figsize=(18,5))
plt.plot(df[['Natural_gas,_US','Shrimps,_Mexican','Barley','Soybean_oil','Platinum','Nickel','Aluminum','Tobacco,_US_import_u.v.','Orange','Sugar,_EU']])
```

```python
important_df=df[['Natural_gas,_US','Shrimps,_Mexican','Barley','Soybean_oil','Platinum','Nickel','Aluminum','Tobacco,_US_import_u.v.','Orange','Sugar,_EU']]
```
