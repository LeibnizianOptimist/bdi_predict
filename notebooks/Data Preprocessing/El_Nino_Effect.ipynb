```python
plt.axhline(y=0, color='black', linestyle='-')
import numpy as np
line=0
y=nino['Temperature']
supper = np.ma.masked_where(y < line, y)
slower = np.ma.masked_where(y > line, y)
X=nino.index
plt.plot(X,supper,'r',X,slower,'b')
```
