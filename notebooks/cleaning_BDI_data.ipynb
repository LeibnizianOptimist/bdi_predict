```python
import pandas as pd
from datetime import datetime as dt
```

```python
# load data from investing.com
df = pd.read_csv("../raw_data/BDI/Weekly Baltic Dry Index Historical Data.csv")
```

```python
# change data type of price string to float
df["Price"] = df["Price"].str.replace(',', '')
df["Price"] = df["Price"].astype("float")

df["Open"] = df["Open"].str.replace(',', '')
df["Open"] = df["Open"].astype("float")

df["High"] = df["High"].str.replace(',', '')
df["High"] = df["High"].astype("float")

df["Low"] = df["Low"].str.replace(',', '')
df["Low"] = df["Low"].astype("float")
```

```python
# change data type of Date string to float
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["price_avg"] = (df["Open"] + df["High"] + df["Low"]) / 3
```
